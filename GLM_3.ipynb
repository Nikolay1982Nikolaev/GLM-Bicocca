##  Domanda 3.1 Multicollinearità: significato
Se la matrice $(X’X)^{-1}$ non è invertibile oppure ha determinante prossimo allo 0, le stime non esistono
(coefficienti sottoidentificati, poiché non si dispone di sufficiente informazione per stimarli) o non
sono stabili (coefficienti empiricamente sottoidentificati). Tale problema si verifica quando almeno
una delle variabili è correlata linearmente alle altre e quindi si ha multicollinearità.

In questo caso la matrice $X’X$ è singolare e le soluzioni non sono uniche.

**Collinearità perfetta** sussiste quando almeno una variabile esplicativa è una combinazione lineare
perfetta delle altre. Diventa impossibile fare stime uniche dei coefficienti di regressione dal
momento che ci sono infinite combinazioni di coefficienti che assicurano la stessa efficacia predittiva
al modello. La perfetta collinearità viola le proprietà del modello lineare classico solitamente per un
errore nella definizione dei regressori o per una stranezza nei dati. Occorre in questo caso modificare
l’insieme dei regressori.

**Imperfetta collinearità** quando 2 o più regressori sono fortemente correlati il determinante della matrice dei coefficienti tende a 0

La presenza di collinearità imperfetta implica che uno o più dei coefficienti di regressione sarà stimato in modo impreciso. Il coefficiente di $x_1$ è l’effetto di $x_1$ tenendo costante $x_2$; ma se $x_1$ e $x_2$
sono altamente correlati, vi è una ridottissima variazione in $x_1$ quando $x_2$ è mantenuta costante,
perciò i dati non contengono molte informazioni su ciò che accade quando $x_1$ cambia e $x_2$ no.
In questo caso, inoltre, la varianza dello stimatore OLS del coefficiente di $x_1$ sarà grande. La
collinearità imperfetta genera grandi errori standard per uno o più dei coefficienti OLS.

Ciò comporta un aumento della varianza delle stime dei coefficienti. Infatti, essendo la varianza delle
stime OLS in cui appare l’inversa di $X’X$

$$Var[\hat{B}]= \sigma^2 (X' X)^{-1}; (1)$$

e ricordando che l’inversa di una matrice $X’X$ è
$$(X’X’)^{-1} = (1 /det (X’X)) (2)$$

in cui il cofattore di ogni elemento è dato dal determinante della sottomatrice in cui è stato eliminata
riga e colonna relativa a quell’elemento si vede che se il determinante è molto piccolo gli elementi
dell’inversa tendono a diventare molto grandi. Ciò porta a conseguenze molto pesanti:

- 1) L’aumento della varianza delle stime dei coefficienti: al variare del campione le stime non sono
stabili e variano moltissimo. Il risultato campionario è quindi molto meno attendibile. Infatti,
considerando la t di Student, se il denominatore (errore standard) aumenta la regione di
accettazione dell’ipotesi nulla si allarga ed aumenta la probabilità che un parametro
statisticamente significativo non sia considerato tale. Analogo ragionamento può essere fatto per
il test F sui singoli parametri essendo $F = t"$
- 2) Se ho due variabili fortemente correlate, la prima spiega ad es. l’80% della varianza dell’outcome
e se aggiungo la seconda, l’incremento di $R^2$ è inferiore all’incremento che avrei aggiungendo una
seconda variabile debolmente correlata con la prima. Ciò avviene anche se nel primo caso la
seconda variabile immessa ha un potere esplicativo molto più elevato che nel secondo caso. Quindi,
siccome le due variabili hanno molta varianza in comune, non posso dire quale delle due è più
influente rispetto all’outcome (è come se le due variabili fossero interscambiabili).

L’approssimazione lineare non è sempre la migliore, ad esempio per via di assunti teorici. Tuttavia, il modello li regressione multipla può gestire funzioni non lineare di una o più covariate.

Se una relazione tra y ed X è non lineare l’effetto su y di una variazione di X dipende dal valore di X,
poiché l’effetto marginale di X non è costante. Una regressione lineare è mal specificata: la forma
funzionale è errata e lo stimatore dell’effetto su y di X non è corretto nemmeno sulla media. Occorre
comunque verificare le condizioni di sfericità degli errori.

image.png

Quando ci si rende conto che la funzione interpolante non può essere lineare occorre procedere alla
trasformazione delle variabili che avviene generalmente utilizzando la funzione logaritmica, la
funzione inversa oppure la radice quadrata o altre funzioni semplici. Una volta trasformate le
variabili si stima il modello di regressione lineare multipla come precedentemente illustrato.
La necessità di trasformare le variabili è dovuta al fatto che:
- (i) con le variabili trasformate le violazioni delle ipotesi del modello sono meno evidenti,
- (ii) l’interpretazione delle variabili è più interessante nel contesto analizzato,
- (iii) l’influenza dei valori anomali è molto ridotta sui dati trasformati,
- (iv) le previsioni del modello stimato con le variabili trasformate risultano più accurate.

Si possono trasformare tutte le variabili oppure solo la variabile risposta oppure le variabili
esplicative. Tuttavia, è bene considerare che la trasformazione della variabile risposta influenza la
sua distribuzione, la varianza e la forma funzionale che la lega alle covariate. Pertanto, la
trasformazione della risposta costituisce un rimedio quando si sospetta che possano essere violate
l’ipotesi di normalità degli errori oppure di omoschedasticità. Diversamente quando si trasformano
le variabili esplicative questo non comporta modificazioni della distribuzione condizionata (alle
covariate) della variabile risposta.

Il modello di regressione può anche essere specificato rispetto agli scarti in media delle variabili
esplicative. Ad esempio, se si considera il piano di regressione possono essere inserite le seguenti
due covariate $xʹ = (x − \bar{x} ̄)$ e $xʹ = (x − \bar{x} ̄)$. In questo caso il modello si riferisce ai valori attesi della
risposta rispetto a valori delle variabili esplicative centrati rispetto alla loro media
