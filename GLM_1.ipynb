Si debbono innanzitutto definire i concetti di:

- **Unità statistica**: è il supporto (persona, oggetto, essere animato) su cui si osservano i fenomeni oggetto di studio;

- **Collettivo statistico o popolazione**: rappresenta l’insieme di unità` statistiche su cui si effettua la rilevazione dei fenomeni oggetto di studio. Pertanto, nello studio sperimentale sull’asma dei topi il collettivo statistico è la popolazione dei topi considerata nel suo complesso. Potenzialmente il collettivo è infinito quando è costituito da un numero di unità il cui ammontare non è precisamente
rilevabile;
- **Carattere statistico**: definisce il fenomeno rilevato sulle unità statistiche di riferimento, Ogni carattere è presente su ogni unità statistiche attraverso specifiche modalità.
I caratteri di stato sono tali quando la modalità del carattere è rilevata in un preciso istante: per esempio i problemi respiratori, il peso di una persona rilevati in un certo istante di tempo, oppure il numero dei figli. I caratteri di movimento sono quelli la cui modalità si riferisce ad un intervallo di
tempo, per esempio il tempo di reazione, il tempo di assorbimento di un farmaco.
Alcune tipologie di caratteri sono sempre invarianti nel tempo o definitivamente invariabili nel tempo come il luogo di nascita per l’individuo. I caratteri variabili nel tempo sono quelli in cui il
tempo è fondamentale per la rilevazione come la durata di una corsa ciclistica.
    - I caratteri quantitativi discreti sono quelli i cui valori indicano una quantità misurata hanno valori numerici interi che derivano da operazioni di conteggio come il numero dei prodotti acquistati su Amazon.
    - I caratteri quantitativi continui derivano da processi di misurazione e dipendono dall’accuratezza degli strumenti usati: a questo scopo occorre definire la loro unità di misura (es. decimi di secondo).

Le variabili qualitative possono essere ordinali, come ad esempio, il rating (punteggio) assegnato dalla valutazione sulla soddisfazione di un servizio, o nominali, come ad esempio il sì o il no espresso al quesito referendario, o l’indicazione del colore preferito.

La statistica si propone come un insieme di metodi che permettono di studiare i fenomeni collettivi, pertanto, si applica a tutte le scienze volte a studiare un evento che coinvolge un collettivo (Leti and Cerbara, 2009). Lo statistico è colui che affronta con metodo scientifico lo studio dei fenomeni collettivi che

I modelli analizzano la relazione fra variabili in modo semplificato e quindi constano di:
- Formulazione di teorie e ipotesi inerenti il fenomeno.
- Individuazione di variabili che esprimono le relazioni inerenti il fenomeno in termini empirici o di causa-effetto tra esse.
- Analisi capaci di interpretare teorie e ipotesi e i nessi tra le variabili.
- Procedure atte a semplificare le analisi stesse.
- Teoremi della statistico-matematici atti a rendere le metodologie più precise e accurate e

Nello studio dei fenomeni collettivi occorre:
- definire il carattere di interesse: ad esempio l’asma;
- individuare la collettività ritenuta di interesse per la conoscenza del fenomeno oggetto di studio: ad esempio i pazienti di un certo ospedale;

La differenza sostanziale tra un modello Matematico e un modello Statistico consiste nel fatto che:
- un modello matematico è deterministico e approssima in maniera esatta i dati. Esplicitano una relazione tra un carattere o variabile dipendente e caratteri o variabili esplicative attraverso una funzione f:

$$y = f(x_1, x_2...x_p) (1)$$

Un modello statistico è invece caratterizzato dalla presenza di errori all’interno della relazione esplicitata
$$y=f(x_1, x_2,...x_p)+ e; (2)$$

che dipendono da:
- a-  Variazioni Individuali
- b-  Componenti sistematiche non incluse nel modello
- c-  Errori di misura

Un modello statistico viene quindi scelto analizzando l’entità dell’errore secondo un trade off tra adattamento ai dati e parsimonia nel numero di parametri, perciò,secondo l’assunto di Popper, ogni modello è intrinsecamente sbagliato. Il modello statistico si serve quindi di variabili per descrivere
le associazioni o causazioni postulate in un fenomeno. La statistica fornisce sia i metodi sia i modelli per l’analisi dei dati rilevati tramite studi sperimentali oppure osservazionali. dell’altra. Si ha invece causazione quando una variazione della prima produce la variazione della
seconda con alcune fondamentali differenze rispetto alla covariazione:

1. la direzionalità: nella causazione esiste una causa e un effetto e il variare di una variabile precede

Il valore immobiliare, ad esempio, potrebbe essere la variabile risposta in un modello in cui si postula

tra di loro, e a valutare l’utilizzo di una semplice funzione lineare per sintetizzare quest’associazione.

L’indice che esprime in termini numerici la direzione e la magnitudine dell’associazione lineare tra
due caratteri quantitativi è il coefficiente di correlazione di Bravais-Pearson. È un indice simmetrico,
considera le variabili nello stesso piano, senza distinguere tra variabile risposta o esplicativa. Il segno
positivo indica correlazione positiva o concordanza: i due caratteri tendono a crescere o decrescere
nella stessa direzione. Il coefficiente di correlazione campionaria si ottiene dividendo la covarianza
osservata tra i due caratteri per il prodotto delle rispettive deviazioni standard. Il campo di
variazione dell’indice è compreso nell’intervallo [−1, 1]. Per i valori estremi si parla di una perfetta associazione lineare tra i due caratteri coinvolti. Un valore zero implica che al crescere o al diminuire
di un carattere non sia associata alcuna tendenza di crescita o decrescita per l’altro carattere e
viceversa. Esprime un’interdipendenza quando i due caratteri non sono legati in senso temporale,
esprime una dipendenza, sempre lineare, se la realizzazione di uno dei due caratteri precede in
senso temporale la realizzazione dell’altro. Un valore pari a 0.85 riferito ad esempio alla correlazione
calcolata tra il valore stimato per un immobile da un’agenzia immobiliare e la superficie complessiva
dello stesso indica una forte correlazione lineare tra i due caratteri che variano congiuntamente
nella stessa direzione, ovvero a fronte di un incremento unitario della superficie c’è un incremento
del valore, e viceversa.

associate.

## Domanda 1.6 Ipotesi modello lineare classico
Nella specificazione di un modello lineare, poiché la relazione non è deterministica ma dipendente
da variazioni casuali del termine di errore (ei) che è caratterizzato da una distribuzione casuale.
Esiste una componente stocastica dell’errore (ei), che è una determinazione campionaria della
variabile casuale non osservabile che descrive l’errore che si commette al variare del campione
spiegando la variabile y tramite le X.
Sia
$$y = X\beta + \epsilon (74)$$

Dove:
- y rappresenta il vettore dei valori della variabile casuale dipendente
- X rappresenta la matrice costituita dall’insieme ordinato delle n osservazioni sulle k variabili
esplicative
- b rappresenta il vettore dei parametri ignoti da stimare
- e rappresenta il vettore degli errori casuali non osservabili

L’equazione (74) con le assunzioni su ei e la sua distribuzione casuale è chiamata modello statistico
lineare.
Nella specificazione, pertanto, del modello Lineare Classico, dobbiamo tenere in considerazioni 6
assunzioni, senza la validità delle quali, il modello specificato potrebbe non risultare valido.
1. Linearità
2. Non sistematicità degli errori
3. Sfericità degli errori
4. Non stocasticità delle variabili esplicative
5. Non collinearità delle variabili esplicative
6. Numerosità della popolazione
