La differenza sostanziale tra un modello Matematico e un modello Statistico consiste nel fatto che:
- un modello matematico è deterministico e approssima in maniera esatta i dati. Esplicitano una relazione tra un carattere o variabile dipendente e caratteri o variabili esplicative attraverso una funzione f:

$$y = f(x_1, x_2...x_p) (1)$$

Un modello statistico è invece caratterizzato dalla presenza di errori all’interno della relazione esplicitata
$$y=f(x_1, x_2,...x_p)+ e; (2)$$

che dipendono da:
- a-  Variazioni Individuali
- b-  Componenti sistematiche non incluse nel modello
- c-  Errori di misura

Un modello statistico viene quindi scelto analizzando l’entità dell’errore secondo un trade off tra adattamento ai dati e parsimonia nel numero di parametri, perciò,secondo l’assunto di Popper, ogni modello è intrinsecamente sbagliato. Il modello statistico si serve quindi di variabili per descrivere

L’indice che esprime in termini numerici la direzione e la magnitudine dell’associazione lineare tra
due caratteri quantitativi è il coefficiente di correlazione di Bravais-Pearson. È un indice simmetrico,
considera le variabili nello stesso piano, senza distinguere tra variabile risposta o esplicativa. Il segno
positivo indica correlazione positiva o concordanza: i due caratteri tendono a crescere o decrescere
nella stessa direzione. Il coefficiente di correlazione campionaria si ottiene dividendo la covarianza
osservata tra i due caratteri per il prodotto delle rispettive deviazioni standard. Il campo di
variazione dell’indice è compreso nell’intervallo [−1, 1]. Per i valori estremi si parla di una perfetta associazione lineare tra i due caratteri coinvolti. Un valore zero implica che al crescere o al diminuire
di un carattere non sia associata alcuna tendenza di crescita o decrescita per l’altro carattere e
viceversa. Esprime un’interdipendenza quando i due caratteri non sono legati in senso temporale,
esprime una dipendenza, sempre lineare, se la realizzazione di uno dei due caratteri precede in
senso temporale la realizzazione dell’altro. Un valore pari a 0.85 riferito ad esempio alla correlazione
calcolata tra il valore stimato per un immobile da un’agenzia immobiliare e la superficie complessiva
dello stesso indica una forte correlazione lineare tra i due caratteri che variano congiuntamente
nella stessa direzione, ovvero a fronte di un incremento unitario della superficie c’è un incremento
del valore, e viceversa.

### 1.Linearita:
Questa ipotesi affronta la forma funzionale del modello. Un modello di regressione è lineare quando
tutti i termini nel modello sono o la costante o un parametro moltiplicato per una variabile
indipendente. Data la seguente specifica del modello, $y= \beta_0 + \beta_1x_1 + \beta_2x_2 + ...+ \beta_kx_k + u_1$ dove i parametri (b) sono le stime da OLS e l'errore (u) e' causale.

La caratteristica che definisce la regressione lineare è questa forma funzionale dei parametri
piuttosto che la capacità di modellare la curvatura. I modelli lineari possono modellare la curvatura includendo variabili non lineari come polinomi e trasformando funzioni esponenziali. Per soddisfare
questa ipotesi, il modello correttamente specificato deve adattarsi al modello lineare.



### 2.Non sistematicità degli errori
Il termine dell'errore rappresenta la variazione nella variabile dipendente che le variabili
indipendenti non spiegano. La casualità dovrebbe determinare i valori del termine di errore.
Affinché il modello sia imparziale, il valore medio del termine di errore deve essere uguale a zero.
Si consideri il modello sulla i-esima osservazione
$$y_i = X_i b + e_i (75)$$

Si ripetano le osservazioni su diverse popolazioni o campioni. Ad esempio, ho fatto in un laboratorio
un esperimento in cui ho studiato il nesso tra il peso di un oggetto e la velocità della sua caduta
ripetendo l’esperimento n volte. Individuo un numero k molto elevato di laboratori in ognuno di
quali ripeto n volte lo stesso esperimento. Si considerino i risultati del primo esperimento in ogni
laboratorio $e_{11}, e_{12},...e_{1k}$. Nel loro complesso tali risultati possono definirsi come una variabile
casuale $E_1$ di cui i risultati in ogni laboratorio $e_{11}, e_{12},...e_{1k}$. sono una particolare manifestazione.

Si può quindi calcolare il valore atteso e la varianza della variabile casuale $E_1$. Si considerino i risultati
del secondo esperimento in ogni laboratorio $e_{21},e_{22},...e_{2k}$. Nel loro complesso tali risultati possono
definirsi come una seconda variabile casuale $E_2$ dalla stessa distribuzione di $E_$1 di cui i risultati in ogni
laboratorio $e_{21},e_{22},...e_{2k}$. sono una particolare manifestazione. Analogamente avviene per i risultati
dell’iesimo esperimento in ogni laboratorio $e_{i1},e_{i2},...e_{ik}$. Nel loro complesso tali risultati possono
definirsi come una variabile casuale $E_i$ dalla stessa distribuzione di $E_1$ e $E_2$ di cui i risultati in ogni
laboratorio $e_{i1}, e_{i2},...e_{ik}$. sono una particolare manifestazione. Si hanno quindi nel complesso n
variabili casuali $E_1, E_2,..., E_i,...E_n$ dalla identica distribuzione. Si può ricavare oltre che il valore atteso
e la varianza di ognuna di esse e la correlazione fra variabili casuali diverse $E_i$ e $E_j$.
Per ciò che concerne il valore atteso nell’ipotesi di non sistematicità degli errori
$$E[E_i]= 0 ; (76)$$
e quindi $$e[E|X]= 0; (77)$$
e conseguentemente:
$$E[y|X]= Xb; (78)$$
$$E(y)= E[Xb+e]= E[Xb]+ E[E]= Xb; (79)$$

SE l'errore medi $E[E_i]$ fosse uguale, ad esempio, a +7 (diverso da zero), questo indicherebbe che il

modello stimato sia inadeguato perché non è corretto in media.

Pertanto, nella stima di un modello si vuole che il valore atteso dell'errore sia uguale a zero.
Altrimenti, se il valore atteso fosse +7 anziché zero, la parte del termine di errore sarebbe
prevedibile e dovremmo aggiungere queste informazioni al modello di regressione stesso, mentre
la specificazione del modello implica che solo gli errori casuali rimasti finiscano all’interno del
termine dell'errore.

Di questa ipotesi, non è necessario preoccuparsi quando si include la costante nel modello di
regressione perché si impone che la media dei residui sia uguale zero. Inoltre il termine di errore e
le covariate debbono essere indipendenti.

### 4.Non stocasticità delle variabili esplicative
Se una variabile indipendente fosse correlata al termine dell'errore, questo ci impedirebbe di usare
la variabile indipendente per prevedere il termine dell'errore, andando così a violare l’assunzione
che il termine di errore sia un errore casuale imprevedibile. Dobbiamo trovare un modo per
incorporare tali informazioni nel modello di regressione stesso. Questa ipotesi viene anche definita
come esogeneità. Quando esiste questo tipo di correlazione, c'è l'endogeneità.

Le violazioni di questa ipotesi possono verificarsi perché vi è simultaneità tra le variabili indipendenti
e dipendenti, la deviazione della variabile omessa o l'errore di misurazione nelle variabili
indipendenti. La violazione di questa ipotesi condiziona la stima del coefficiente.

Le $x_j$ sono fisse (j=1,2..p)

### 5.Non collinearità delle variabili esplicative
Premettendo che la correlazione perfetta si verifica quando due variabili hanno un coefficiente di
correlazione di Pearson di +1 o -1, si ha che in questo caso quando una delle variabili cambia, anche
l'altra variabile cambia di una proporzione completamente fissa, pertanto, le due variabili si
muovono all'unisono. Una correlazione perfetta suggerisce che due variabili sono forme diverse
della stessa variabile. Ad esempio, la temperatura in gradi Fahrenheit e Celsius ha una perfetta
correlazione positiva (+1). La correlazione perfetta può essere anche tra una variabile e una
combinazione lineare di alcuna delle altre o di tutte le altre.

I minimi quadrati ordinari non possono distinguere una variabile dall'altra quando sono
perfettamente correlate. Questo dipende dal fatto che la matrice X non sarebbe a rango pieno, e
quindi la matrice $X’X$ necessaria per il calcolo dei coefficienti sarebbe singolare e quindi non
invertibile.

Quando avvenisse correlazione perfetta tra alcune delle variabili, i software non sarebbero in grado
di stimare il modello.

la matrice X ha rango pieno, pari a p+1 ovvero il numero delle covariate, costante inclusa:

Ne consegue che la matrice (X'X) non e' singolare.

distribuito in modo normale. QUindi:

Per n grande (maggiore di 25) per il teorema del limite centrale la distribuzione di $\epsilon_i$ e quindi di y e $\hat{B}$ tendono alla normalità. In caso di non normalità di $\epsilon_i$
 con n piccolo (minore di 25) determinare la

Il metodo della massima verosimiglianza è un altro metodo impiegato per la stima dei parametri del
modello di regressione lineare multipla. La funzione di verosimiglianza (L) o il logaritmo della
verosimiglianza l (log- verosimiglianza) descrive il supporto che i dati osservati attribuiscono ai

che è stato definito lo spazio parametrico in relazione alla funzione di interesse f (obiettivo).
Consente di scegliere quei valori dei parametri che risultano maggiormente plausibili alla luce dei
dati osservati. Pertanto, la stima di massima verosimiglianza è quella che in base al modello ed ai
dati disponibili offre maggiore evidenza empirica. Per approfondimenti si rimanda ad Azzalini (2004).
Si dimostra che gli stimatori dei parametri che si ricavano con questo metodo coincidono con quelli
ottenuti con il metodo dei minimi quadrati. Si richiamano le seguenti caratteristiche della funzione
di verosimiglianza considerando il caso uniparametrico. Si dispone di un vettore $y = (y_1...y_n)$ di n osservazioni indipendenti ed identicamente distribuite, $f(y; \theta)$ e' una dunzione dove $\theta$ rappresenta il parametro igoto $\theta \in \Theta \subset R$ R. Nell’approccio frequentista all’inferenza si pone l’accento sul fatto che
i dati sono informativi sul parametro in quanto realizzazioni di una variabile aleatoria Y.
Nell’inferenza Bayesiana invece il parametro θ è descritto mediante una distribuzione di densità
(probabilità). La funzione di verosimiglianza è riferita alla probabilità di ottenere la n-upla dei valori
realmente osservati come realizzazioni del vettore aleatorio Y. Si scrive pertanto in funzione del
parametro e si scrive
$$L(\theta)= \prod_{i=1}^n f(y_i; \theta);(96)$$ Il logaritmo naturale della funzione di verosimiglianza è una funzione più agevole da massimizzare

per ricavare la stima puntuale e stima intervallare del parametro. Lo stimatore di massima log-
verosimiglianza si denota con θ ed è quell’unico valore del parametro θ in corrispondenza del

quale la funzione di log-verosimiglianza raggiunge il punto di massimo $\hat{l}(\theta)$.

Sulla base della teoria del test di ipotesi, la distribuzione di un parametro studiata sotto l’ipotesi
nulla divide in due parti l’area sottesa a questa distribuzione sulla base di una quantità pivotale:
- l’area di accettazione del test che comprende i valori associati alla quasi totalità delle
probabilità cumulate della distribuzione
- l’area di rifiuto che comprende i valori associati alle code della distribuzione.

Nei test inerenti i parametri del modello lineare se il valore campionario βj trovato cade in una delle
due code della distribuzione, allora si è verificato uno dei seguenti casi.
- L’ipotesi nulla è vera e si è verificato un evento raro
- L’ipotesi nulla è falsa

Si sceglie di solito la seconda interpretazione e quindi si rifiuta $H_0$ al livello di significatività del 100α%
se il valore campionario $β_j$ trovato cade in una delle due code della distribuzione e non rifiutare $H_0$
negli altri casi.

Se il valore del parametro ricavato dal campione ricade nell’area di accettazione si accetta l'ipotesi
nulla legata al valore del campione; alternativamente si accetta l’ipotesi alternativa. Procedendo in
questo modo si minimizza l’errore di prima specie cioè la probabilità di rifiutare l'ipotesi nulla
quando è vera. Solitamente nel modello lineare tale ipotesi nulla è l’ipotesi che il parametro sia
nullo, ovvero che non ci siano legami tra regressore e variabile dipendente. E quindi si minimizza
l’ipotesi di sostenere la tesi che ci siano legami tra regressore e variabile dipendente quando in realtà
non ci sono.

Perciò per non rifiutare l’ipotesi che vi siano legami tra variabile dipendente ed esplicativa il valore
campionario deve essere molto diverso da zero, maggiore in valore assoluto della quantità pivotale
e cadere nelle code della distribuzione. Il test è molto cautelativo rispetto alla ipotesi di non escludere l’esistenza di legami. Quindi, nel modello lineare, l’ipotesi nulla che si ha interesse a
verificare è se il parametro legato ad una determinata variabile inserita nel modello sia nullo.

### 5.Intervallo di confidenza
Per la costruzione di un intervallo di confidenza per il parametro della regressione al 95%, significa
che se dovessimo prelevare 100 campioni diversi e calcolare un intervallo di confidenza del 95% per
ciascun campione, allora circa 95 dei 100 intervalli di confidenza conterranno il vero valore medio
$(μ)$. In pratica, tuttavia, selezioniamo un campione casuale e generiamo un intervallo di confidenza,
che può contenere o meno la media vera. L'intervallo osservato può sovrastimare o sottostimare $μ$.
Di conseguenza, l'IC del 95% è l'intervallo probabile del parametro vero e sconosciuto.

L'intervallo di confidenza non riflette la variabilità del parametro sconosciuto. Piuttosto, riflette la
quantità di errori casuali nel campione e fornisce un intervallo di valori che potrebbero includere il
parametro sconosciuto. Un altro modo di pensare a un intervallo di confidenza è che si tratta
dell'intervallo di valori probabili del parametro (definito come la stima puntuale + margine di errore)
con un determinato livello di confidenza (che è simile a una probabilità). Se conoscessimo il vero
valore del parametro, allora sui campioni generati potremmo scartare quelli che non lo contengono,
ma non sapendo quale sia il vero valore del parametro non possiamo procedere in questo modo.

### 6.Relazione fra test di ipotesi e intervalli di confidenza
Rifiutiamo $H_0$ se il p-value è piccolo e quindi c’è solo una piccola probabilità che il parametro nel
campione ha un valore estremo anche se $H_0$ è vero mentre un intervallo di confidenza al 95%
fornisce un intervallo di valori plausibili per il coefficiente angolare in cui il 95% degli intervalli
comprende il vero valore del coefficiente angolare stesso.

Poiché il valore atteso della derivata del logaritmo funzione di verosimiglianza rispetto a $β,σ^2$ si
dimostra essere nullo, il valore atteso della derivata seconda del logaritmo funzione di

Tanto minore sarà la variabilità contenuta in questa matrice di varianze covarianze tanto minore
sarà la quantità di informazione contenuta dalla variabile osservabile $ε = y-Xβ$, concernente i
parametri non osservabili $β,σ^2$ , da cui dipende la distribuzione di probabilità di $ε = y-Xβ$. (misura
della curvatura della verosimiglianza in corrispondenza della stima di massima verosimiglianza) per
$β,σ^2$

Il teorema di Cramer Rao afferma che lo stimatore a minima varianza fra tutti gli stimatori anche
non lineari è quello la cui varianza è uguale alla inversa della matrice di informazione. Si dimostra,
infine, che la matrice di varianze covarianze degli stimatori di massima verosimiglianza di $β,σ^2$
coincide con l’inversa della matrice di informazione di Fisher (70) e quindi sono gli stimatori VUE
anche non lineari.
