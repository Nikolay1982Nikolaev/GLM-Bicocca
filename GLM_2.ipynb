## Domanda 2.4 Violazione modello classico: autocorrelazione
Frequentemente i residui sono correlati tra di loro. Così, soprattutto in serie storiche o territoriali, è
ragionevole ipotizzare che esista correlazione fra errori in momenti successivi o territori vicini. Come
nei seguenti casi:
- Modelli longitudinali: che studiano il reddito con autocorrelazione tra osservazioni di periodi
successivi
- Modelli spaziali su livello di produzione in comuni limitrofi
- Modelli di scelte nel consumo di redditi ricchi e poveri
- Modelli di indicatori di gravità malattia nell’invecchiamento con autocorrelazione negli anni
Nel modello classico si suppone che:
$$Cov(e_i; e_j)= E(e_i; e_j)= 0$$
- per ogni $i \ne j; (24)$

Quando questa condizione non è verificata si dice che gli errori sono correlati o che si è in presenza
di correlazione seriale. Con omoschedasticità e correlazione la matrice di correlazione degli errori
diviene

image.png

Sia dato il seguente modello
$$y_i = X_i \hat{B}+ \epsilon_i\#;(26)$$
con errori correlati
$$e_i\#= \rho e_{i-1} +\eta_i;(27) \$$

dove $\eta_i$ sono IID identicamente e indipendentemente distribuiti in modo normale $(0, \sigma_{\eta})$

Pertanto, $e_i^{\#}$ e definito come un processo autoregressivo perche $e_i^{\#}$ e' legato al suo valore ritardato $e_{i-1}^{\#}$ essendo $e^{\#}$ gli errori correlati per distinguerli dagli errori sferici $e$.

Si dice che c’è autocorrelazione se al variare di x c’è fluttuazione dei valori di y con lo stesso segno

## Domanda 2.7 Normalità errori e dimensione del campione
Si sia ricavato dalla popolazione un certo campione $ε_{11},..., ε_{21},..., ε_{n1}$.

Si estraggano un numero infinito di campioni (che indichiamo per convenzione k) adottando uno schema di campionamento casuale di tipo semplice Si considerino i primi elementi di ogni campione
$ε_{11},..., ε_{1i},..., ε_{1k}$ che, nel loro complesso, al variare del campione sono la manifestazione di una
variabile casuale $E_1$.

Si considerino i secondi elementi di ogni campione $ε_{21},..., ε_{2i},..., ε_{2k}$ che, nel loro complesso, al variare
del campione sono la manifestazione di una variabile casuale $E_2$ dalla stessa distribuzione di $E_1$.

Si considerino gli n-esimi elementi di ogni campione $ε_{n1},..., ε_{ni},..., ε_{nk}$ che, nel loro complesso, al

$E_1, E_2$.

nella popolazione è normale $E N(0,σ2)$. Perciò ognuna delle n variabili casuali campionarie $E_1, E_2,...,
E_i,...E_n$ ha distribuzione identica e normale.

normalità degli errori.




## Domanda 2.8 Conseguenze della non normalità degli errori
1. Se gli errori non sono normali i parametri b, espressi come combinazione lineare degli errori, non
sono più distribuiti normalmente e anche le loro stime B.
2. Non è più possibile ricavare test basati sulla normale standardizzata, sulla t di Student e sulla F di
Snedecor per i parametri e test sull’adattamento del modello basati sulla F di Snedecor.
3. Non è più possibile ricavare intervalli di confidenza per i parametri basati sulla normale
standardizzata (quando si conosce la varianza della popolazione), sulla t di Student.
4. Le stime OLS non coincidono con le stime ML ottenute attraverso il metodo della massima
verosimiglianza perché ciò vale solo con la normalità.
5. Si hanno problema computazionali perché le stime ottenute con i pacchetti statistici (R, Sas) sono
basate su verosimiglianza.
6. Per gli stimatori OLS, non essendo più equivalenti agli stimatori di massima verosimiglianza, non
vale più il teorema di Cramer-Rao Quindi gli stimatori OLS non sono più gli stimatori corretti a
minima varianza fra tutti gli stimatori corretti (VUE). Tuttavia, gli stimatori OLS rimangono corretti

### 3 Tests non parametrici
Si dicono non parametrici (o indipendenti dalle distribuzioni) i test che non necessitano di particolari
condizioni per essere applicati
Non è necessario che il campione contenga più di un certo numero di elementi e che le grandezze
confrontate seguano una certa distribuzione. Non hanno come scopo la stima dei parametri della
popolazione quanto il tipo di distribuzione.

Sono particolarmente adatti a confrontare campioni molto piccoli e/o non seguenti alcuna distribuzione nota.

- a) Test di Shapiro-Wilk

Si consideri la grandezza
$$W = \frac{\sum \beta_i \epsilon_i}{\sum \epsilon_i^2};(63)$$

avente al numeratore una combinazione lineare dei residui in ordine crescente e al numeratore la
varianza campionaria. I pesi $\beta_3$ sono generati da media varianza e covarianza presi da una
distribuzione normale. W è compreso tra 0 e 1, estremi che corrispondono rispettivamente al rifiuto
e all’accettazione dell’ipotesi di normalità. Infatti, le costanti sono costruite in modo tale che la
statistica W tenda a 1 se i residui provengono da un Normale.

$H_0$: W cade nella regione di accettazione. Non si respinge l’ipotesi di normalità

$H_1$: distribuzione non normale se il p value corrispondente al valore di W è superiore al livello di
significatività $α$

Vi è da aggiungere che W è fortemente asimmetrico e suoi valori elevati possono portare ugualmente al rifiuto dell’ipotesi di normalità.

- b) Test Kolmogorov Smirnov

Si suddivide l’intervallo di variazione in classi di frequenza di uguale ampiezza. Ad ogni classe si
attribuiscono le frequenze cumulative della distribuzione empirica da testare e della normale.

Si procede alla differenza fra le frequenze classe per classe.

Si considera la differenza maggiore in valore assoluto D Si ricorre alle apposite tavole per la D. Se D
supera il valore critico tabulato al livello di significatività prescelto, la differenza fra i campioni è
significativa e l’ipotesi di normalità è respinta

### c) Test di Skewness (test di asimmetria) Il test di skewness è essenzialmente un test direzionale.
basato sul fatto che la distribuzione della normale è simmetrica e quindi sull’indice di simmetria.
Sebbene il test di asimmetria sia un potente test di direzionalità, bisogna assicurarsi che il campione
sia sufficientemente grande perché il test non riconosce la dipartita dalla distribuzione normale
quando il campione è piccolo. Per questo a partire dai dati il valore dell’indice è spesso ottenuto da
una simulazione con 1000 iterazioni

$$S= \frac{(E[X-\mu]^3)^2}{E[X-\mu]^2}^3; (64)$$

Rigettando $H_0$ si rigetta la normalità. Non rigettando si dice solo che la distribuzione è simmetrica
(non è detto sia normale)



### d) Test della Kurtosis
Anche il test della Kurtosis, così come quello di asimmetria, è un test direzionale che verifica se la
distribuzione empirica ha la stessa curtosi della normale vale a dire lo stesso modo di crescere e
descrescere della normale, verificando se sia più “gonfia" o più sottile di quanto sia la normale.
Perciò è basato sull’indice di curtosi

$$K = \frac{E[X-\mu]^4}{(E[X-\mu]^2)^2}$$

4^ momento intorno alla media/ Quadrato della varianza
Il test è dato alla distribuzione di K-3 che sotto ipotesi di normalità ha valore atteso E(K-3)=0
Rigettando H0 si rigetta la normalità. Non rigettando si dice solo che la distribuzione ha curtosi 3
(non è detto sia normale).
